## Task 1 – Email Processing

```python
def process_email(email):
    if email.count("@") != 1:
        return "Invalid email"

    username, rest = email.split("@")

    if "." not in rest:
        return "Invalid email"

    domain = rest.rsplit(".", 1)[0]

    if email.endswith(".com"):
        domain_type = "Commercial Domain"
    elif email.endswith(".edu"):
        domain_type = "Educational Domain"
    else:
        domain_type = "Other Domain"

    return {
        "username": username,
        "domain": domain,
        "domain_type": domain_type
    }

process_email("Amit_ml@gmail.edu")
```

## Task 2 – Decode Message


```python
encoded = "###!!@mocleW EPGTQ!!!6789"

core = "".join(ch for ch in encoded if ch.isalpha() or ch == " ").strip()
first, second = core.split()

decoded = first[::-1] + " " + second
decoded
```

## Task 3 – Decode Message with Vowel Shift


```python
encoded = "&&&**$gnirtS PLIO!!@1234"

core = "".join(ch for ch in encoded if ch.isalpha() or ch == " ").strip()
first, second = core.split()

vowels = {"I": "E", "O": "U"}
second = "".join(vowels.get(ch, ch) for ch in second)

first[::-1] + " " + second
```
