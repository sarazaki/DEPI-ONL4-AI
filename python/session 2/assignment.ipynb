17. Replace the last value of each tuple in a list

- Sample list:
- [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
- Expected output:
- [(10, 20, 100), (40, 50, 100), (70, 80, 100)]

```python
data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

# TODO: replace last value with 100
new_list = []
# your code here
for t in data:
    temp = list(t)
    temp[-1] = 100
    new_list.append(tuple(temp))
print(new_list)
# Hint: Convert each tuple to list, edit, convert back.
```
